## Import necessary libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set(style='whitegrid')
```

## Load the Dataset

```python
file_path = 'train.csv'  # Replace with your file path if different
titanic_df = pd.read_csv(file_path)
titanic_df.head()
```

## Data Overview

```python
titanic_df.info()
titanic_df.isnull().sum()
titanic_df.describe()
```

## Data Cleaning

```python
titanic_df['Age'] = titanic_df['Age'].fillna(titanic_df['Age'].median())
titanic_df['Embarked'] = titanic_df['Embarked'].fillna(titanic_df['Embarked'].mode()[0])
titanic_df.drop(columns=['Cabin'], errors='ignore', inplace=True)
titanic_df.isnull().sum()
```

## Exploratory Data Analysis (EDA)

## Univariate Analysis

```python
sns.countplot(x='Survived', data=titanic_df)
plt.title('Distribution of Survival')
plt.show()
sns.countplot(x='Pclass', data=titanic_df)
plt.title('Passenger Class Distribution')
plt.show()
sns.histplot(titanic_df['Age'], bins=30, kde=True)
plt.title('Age Distribution')
plt.show()
```

## Bivariate Analysis

```python
sns.countplot(x='Survived', data=titanic_df)
plt.title('Distribution of Survival')
plt.show()

sns.countplot(x='Pclass', data=titanic_df)
plt.title('Passenger Class Distribution')
plt.show()

sns.histplot(titanic_df['Age'], bins=30, kde=True)
plt.title('Age Distribution')
plt.show()
```

## Multivariate Analysis

```python
numeric_cols = titanic_df.select_dtypes(include=['float64', 'int64'])
plt.figure(figsize=(10, 8))
sns.heatmap(numeric_cols.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix')
plt.show()
```
